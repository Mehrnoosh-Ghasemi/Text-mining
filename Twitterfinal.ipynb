```python
import re
def remove_emoji(string):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r' ', string)
```

## Topic Modeling using NMF

```python
SENT = []
for list in list2:
    string = ' '.join([str(item) for item in list])
    sent = " ".join(w for w in nltk.wordpunct_tokenize(string) if w.lower() in words or w.isalpha() ).encode('ascii', 'ignore').decode('utf-8')
    sent1 = " ".join(w for w in nltk.wordpunct_tokenize(sent) if len(w)>3)
    SENT.append(sent1 )

SENT_DF = pd.DataFrame(SENT)
SENT_DF
```

```python
import pandas as pd
import numpy as np
```

```python
reviews_datasets =SENT_DF
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(max_df=0.8, min_df=2, stop_words='english')
doc_term_matrix = tfidf_vect.fit_transform(reviews_datasets[0].values.astype('U'))
```

```python
from sklearn.decomposition import NMF

nmf = NMF(n_components=17, random_state=42)
nmf.fit(doc_term_matrix )
```

```python
import random

for i in range(10):
    random_id = random.randint(0,len(tfidf_vect.get_feature_names()))
    print(tfidf_vect.get_feature_names()[random_id])
```

```python
first_topic = nmf.components_[0]
top_topic_words = first_topic.argsort()[-10:]
```

```python
for i in top_topic_words:
    print(tfidf_vect.get_feature_names()[i])
```

```python
for i,topic in enumerate(nmf.components_):
    print(f'Top 10 words for topic #{i}:')
    print([tfidf_vect.get_feature_names()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

```python
topic_values = nmf.transform(doc_term_matrix)
reviews_datasets['Topic'] = topic_values.argmax(axis=1)
reviews_datasets
```

```python
mergedDf = new2.merge(reviews_datasets, left_index=True, right_index=True)
mergedDf
```
